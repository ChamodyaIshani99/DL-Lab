```python
# 1️⃣ Clean column name issues (like ï»¿Day of Week)
data.columns = data.columns.str.strip().str.replace('ï»¿', '', regex=False)

# 2️⃣ Clean the values in "Day of Week"
data['Day of Week'] = data['Day of Week'].astype(str).str.strip()

# 3️⃣ Now redefine and apply your function
def add_day(inptx):
    if inptx in ['Friday', 'Thursday']:
        return 'Other'
    else:
        return inptx

data['Day of Week'] = data['Day of Week'].apply(add_day)

# 4️⃣ Finally, check again
print(data['Day of Week'].value_counts())
```
